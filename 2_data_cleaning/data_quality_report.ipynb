# Data Quality Report For Laverie1

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

base_path = '../data/laverie1/' 

# Fichiers Ã  analyser
files = [
    "transactions_jour.csv",
    "transactions_transaction.csv",
    "remplissages.csv",
    "releves.csv",
    "alertes.csv"
]

for file in files:
    print("="*70)
    print(f" Rapport qualitÃ© : {file}")
    print("="*70)

    try:
        df = pd.read_csv(os.path.join(base_path, file), sep=None, engine="python")
    except Exception as e:
        print(f" Erreur de chargement : {e}")
        continue

    display(df.head())
    print("\n Dimensions :", df.shape)
    print("\n Colonnes :", df.columns.tolist())
    print("\n Valeurs manquantes :")
    display(df.isnull().sum())

    print("\nðŸ“Ž Types de donnÃ©es :")
    display(df.dtypes)

    print("\n Statistiques numÃ©riques :")
    display(df.describe())

    # Graphique rapide des valeurs manquantes
    plt.figure(figsize=(10,4))
    sns.heatmap(df.isnull(), cbar=False, cmap="Reds")
    plt.title(f"Valeurs manquantes - {file}")
    plt.show()

    print("\n\n\n")
```
