### Objectif 1.1 : Suivre et prédire le chiffre d’affaires (pour les deux années suivantes)

##### Visualisation chiffre d'affaire journalier 

```python
# 3_eda_visualisation/eda.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement
df = pd.read_csv("../data_cleaned/laverie1/transactions_jour_cleaned.csv", parse_dates=["date"])
df = df.rename(columns=str.lower)

# Plot CA
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="date", y="ca_tot")
plt.title("Chiffre d'affaires journalier")
plt.xlabel("Date")
plt.ylabel("CA total (€)")
plt.grid()
plt.tight_layout()
plt.savefig("graphs/ca_journalier.png")
plt.show()
```
